```python
import pandas as pd
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num

# Assuming you have loaded the data into a DataFrame named "mastercard_data"
# Convert the "Date" column to datetime type if it's not already in datetime format
mastercard_data['Date'] = pd.to_datetime(mastercard_data['Date'])

# Set the "Date" column as the index
mastercard_data.set_index('Date', inplace=True)

# Resample the data to weekly frequency for better candlestick visualization
weekly_data = mastercard_data.resample('W').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last'})

# Convert dates to the format required by candlestick_ohlc
weekly_data.reset_index(inplace=True)
weekly_data['Date'] = weekly_data['Date'].apply(date2num)

# Plot the candlestick chart
plt.figure(figsize=(12, 6))
ax = plt.subplot()
candlestick_ohlc(ax, weekly_data.values, width=0.6, colorup='g', colordown='r')
ax.xaxis_date()
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m-%d'))
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Mastercard Weekly Candlestick Chart')
plt.show()
```

```python
#Moving Averages
plt.figure(figsize=(12, 6))
mastercard_data['MA50'] = mastercard_data['Close'].rolling(window=50).mean()
mastercard_data['MA200'] = mastercard_data['Close'].rolling(window=200).mean()
plt.plot(mastercard_data.index, mastercard_data['Close'], label='Close Price', color='blue')
plt.plot(mastercard_data.index, mastercard_data['MA50'], label='50-day Moving Average', color='orange')
plt.plot(mastercard_data.index, mastercard_data['MA200'], label='200-day Moving Average', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Mastercard Stock Price with Moving Averages')
plt.legend()
plt.show()
```

## 2, Stock Market prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Convert the "Date" column to datetime type if it's not already in datetime format
# Assuming you have the Mastercard stock data in a CSV file, you can load it into a DataFrame as follows:
mastercard_data = pd.read_csv('Mastercard_stock_history.csv')

# Then, you can follow the subsequent steps to convert the "Date" column to datetime type and set it as the index.

mastercard_data['Date'] = pd.to_datetime(mastercard_data['Date'])

# Set the "Date" column as the index
mastercard_data.set_index('Date', inplace=True)

# Sort the data chronologically (if not already sorted)
mastercard_data.sort_index(inplace=True)
```

```python
#Split the Data into Train and Test Sets

#Split the Data into Train and Test Sets:
#To evaluate the performance of our model, we'll split 
#the data into training and testing sets. We'll use historical data for training and try to forecast future stock prices.
# Set the proportion of data to be used for training (e.g., 80%)

train_proportion = 0.8
train_size = int(len(mastercard_data) * train_proportion)
train_data = mastercard_data.iloc[:train_size]
test_data = mastercard_data.iloc[train_size:]
```

Build and Train the ARIMA Model:

We'll use the ARIMA function from statsmodels to build the ARIMA model. The order of the ARIMA model is represented as (p, d, q), where:

p: the number of autoregressive terms (lags)


d: the degree of differencing (to make the time series stationary)


q: the number of moving average terms (lags)

```python
# Choose the order for the ARIMA model based on data analysis and ACF/PACF plots
p, d, q = 1, 1, 1
arima_model = ARIMA(train_data['Close'], order=(p, d, q))
arima_model_fit = arima_model.fit()
```

Make Predictions:


Use the trained ARIMA model to make predictions on the test set.

```python
predictions = arima_model_fit.forecast(steps=len(test_data))
```

Visualize the Results:


Plot the predicted stock prices against the actual prices to visually assess the model's performance.

```python
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data['Close'], label='Training Data', color='blue')
plt.plot(test_data.index, test_data['Close'], label='Actual Prices', color='green')
plt.plot(test_data.index, predictions, label='Predicted Prices', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('ARIMA Model - Stock Price Prediction')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test_data['Close'], predictions))
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
